```python
def roul():
    return np.random.randint(0, 37)
```

### 2

#### 1)

```python
n = 1000000
k = 0

for _ in range(n):
    value = roul()
    if value == 5 or value == 10:
        k += 1
print(f"Статистическая частота: {k / n}")
print(f"Вероятность: {1 / 37 + 1 / 37}")
```

#### 1)

```python
k, n = 0, 10000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
x = a + b + c + d

for i in range(0, n):
    if x[i] == 2:
        k = k + 1
print(k / n)
```

#### 2)

```python
k, n = 0, 10000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
e = np.random.randint(0, 2, n)
f = np.random.randint(0, 2, n)
g = np.random.randint(0, 2, n)
x = a + b + c + d + e + f + g

for i in range(0, n):
    if x[i] == 5:
        k = k + 1
print(k / n)
```

```python
n = 100
r = 0.7
x = np.random.random(n)
y = r * x + (1 - r) * np.random.random(n)

np.sum((x - x.mean()) * (y - y.mean())) / np.sqrt(np.sum((x - x.mean())**2) * np.sum((y - y.mean())**2))
```
